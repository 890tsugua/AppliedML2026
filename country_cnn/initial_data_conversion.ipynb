### Export labels to a csv


```python
train_labels = ds["train"]
```

```python
train_labels = train_labels["label"]
print(train_labels[:5])
```

```python
# Save list to csv
with open("train_labels.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for x in train_labels:
        writer.writerow([x])
```

```python
test_labels = ds["test"]["label"]
with open("test_labels.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for x in test_labels:
        writer.writerow([x])
```

```python
from pathlib import Path

# Path to your folder
root = Path("/Users/august/.cache/huggingface/hub/datasets--deboradum--GeoGuessr-countries/snapshots/2a6258e1e6cde675cdb7366ba125d8456baa78bf/test")

# Get all subfolder names
folder_names = sorted([
    f.name for f in root.iterdir() if f.is_dir()
])

# Map alphabetically to indices
folder_to_idx = {
    name: idx for idx, name in enumerate(folder_names)
}

print(folder_to_idx)
```

```python
# Make index to folder mapping
idx_to_folder = {idx: name for name, idx in folder_to_idx.items()}
print(idx_to_folder)
```

```python
# Save folder_to_idx to csv
with open("country_to_idx.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for key, value in folder_to_idx.items():
        writer.writerow([key, value])
```

### Export images to google drive

```python
# '/Users/august/Library/CloudStorage/GoogleDrive-emmak2911@gmail.com/Mit drev/AppliedMLdata2026'
```

```python
from pathlib import Path
from PIL import Image
from tqdm import tqdm

drive_root = Path(
    "/Users/august/Library/CloudStorage/GoogleDrive-emmak2911@gmail.com/Mit drev/AppliedMLdata2026"
)

idx_to_country = idx_to_folder

target_size = (500, 281)  # width, height

for i, (img, label) in enumerate(tqdm(zip(ds["train"]["image"], train_labels), total=len(train_labels))):
    country = idx_to_country[int(label)]

    country_dir = drive_root / country
    country_dir.mkdir(parents=True, exist_ok=True)

    # Ensure RGB, resize, save
    img = img.convert("RGB")
    img = img.resize(target_size, Image.Resampling.LANCZOS)

    out_path = country_dir / f"{i:08d}.jpg"
    img.save(out_path, quality=90)
```

```python
drive_root = Path(
    "/Users/august/Library/CloudStorage/GoogleDrive-emmak2911@gmail.com/Mit drev/AppliedMLdata2026/test_data"
)

idx_to_country = idx_to_folder

target_size = (500, 281)  # width, height

for i, (img, label) in enumerate(tqdm(zip(ds["test"]["image"], test_labels), total=len(test_labels))):
    country = idx_to_country[int(label)]

    country_dir = drive_root / country
    country_dir.mkdir(parents=True, exist_ok=True)

    # Ensure RGB, resize, save
    img = img.convert("RGB")
    img = img.resize(target_size, Image.Resampling.LANCZOS)

    out_path = country_dir / f"{i:08d}.jpg"
    img.save(out_path, quality=90)
```
